# Policy RAG 실험 분석

- `experiments/` 의 results.json, eval.json 을 불러와 정확도·규칙 검색률 비교
- 실험별 config (chunk_size, top_k 등) 와 메트릭 시각화

```python
import json
from pathlib import Path

ROOT = Path("..")
experiments_dir = ROOT / "experiments"

def load_eval(exp_name: str):
    p = experiments_dir / exp_name / "eval.json"
    if not p.exists():
        return None
    with open(p, encoding="utf-8") as f:
        return json.load(f)

# 예: exp01, exp02 비교
for name in ["exp01_vanilla_llm", "exp02_rag_baseline"]:
    ev = load_eval(name)
    if ev:
        print(f"{name}: accuracy={ev.get('accuracy', 0):.4f}, rule_retrieval_rate={ev.get('rule_retrieval_rate', 0):.4f}")
```
