# Policy RAG 실험 분석

- `experiments/` 의 predictions.json, evaluation.json 로드
- 정확도·규칙 검색률·에러 타입 집계
- `results/tables/`, `results/figures/` 에 저장

```python
import json
from pathlib import Path

ROOT = Path("..")
experiments_dir = ROOT / "experiments"
results_dir = ROOT / "results"

def load_evaluation(exp_path: str):
    """e.g. vanilla/run1, rag_baseline/run1"""
    p = experiments_dir / exp_path / "evaluation.json"
    if not p.exists():
        p = experiments_dir / exp_path / "eval.json"  # fallback
    if not p.exists():
        return None
    with open(p, encoding="utf-8") as f:
        return json.load(f)

for name in ["vanilla/run1", "rag_baseline/run1"]:
    ev = load_evaluation(name)
    if ev:
        print(f"{name}: accuracy={ev.get('accuracy', 0):.4f}, rule_retrieval_rate={ev.get('rule_retrieval_rate', 0):.4f}")
```
